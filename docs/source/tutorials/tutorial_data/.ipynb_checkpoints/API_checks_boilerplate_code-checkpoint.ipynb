### more detailed checks in a shorter time period, e.g. 2 weeks (depending on data, something you can control "by hand")

```python
# filter for more detailed data
detailed_start_date = '2020-02-02'
detailed_end_date = '2020-02-16'
detailed_data = data[(data['date'] >= detailed_start_date) & (data['date'] <= detailed_end_date)]
```

```python
detailed_data
```

```python
# check statistics for detailed data
item_to_check = 'amount' #'turnover'
detailed_data[list(detailed_data.filter(regex=item_to_check))].sum().sum()
```

```python
# check specific plant
plant_to_check = 'Dianthus'
detailed_data.filter(regex=plant_to_check + '|date')
```
